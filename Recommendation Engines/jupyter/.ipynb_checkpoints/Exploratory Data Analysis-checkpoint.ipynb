# Exploring the data before making Review Recommendations for BoardGameGeek.com

For this project, I have three datasets scraped from boardgamegeeks.com. The elite set has reviews from the top 199 most prolific reviewers on site, Frequent has the top 2,473 reviewers. The last set has all users (including elite and frequent) for 193,504 users total. 

The intent, is to create a system for predicting what user u would review a given item i. Rather than using a library, I'm using this project as a chance to explore and implement algorithms from <a href="http://www.nowpublishers.com/article/Details/HCI-009">this paper</a>. 

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline

elite = pd.read_csv('../inputs/boardgame-elite-users.csv')
elite = elite.rename(columns = {'Compiled from boardgamegeek.com by Matt Borthwick':'UserID'})
frequent = pd.read_csv('../inputs/boardgame-frequent-users.csv')
frequent = frequent.rename(columns = {'Compiled from boardgamegeek.com by Matt Borthwick':'UserID'})
titles = pd.read_csv('../inputs/boardgame-titles.csv')
titles = titles.rename(columns={'boardgamegeek.com game ID':'gameID'})
users = pd.read_csv('../inputs/boardgame-users.csv')
users = users.rename(columns = {'Compiled from boardgamegeek.com by Matt Borthwick':'UserID'})
```

Looks like elite is a subset of users from frequent. No need to combine datasets for larger datasets then, you just need to switch to the next size up. 

```python
len(set(frequent.UserID.unique()).intersection(set(elite.UserID.unique())))
```

## Review score Distribution in 'Frequent' dataset

Top graph is with linear axis, bottom graph has a y axis that's log scale. Notice the distribution for each level of review (whole number, half number, fractional number) is directly related. Also notice the distribution isn't quite normal. For my analysis below, it's still close enough that I'm going to assume normal distribution.

```python
fig, (ax1, ax2) = plt.subplots(2,1,figsize=(13,20))
ax2.set_yscale('log')
frequent.rating.hist(bins=250,ax=ax1)
frequent.rating.hist(bins=250,ax=ax2,bottom=0.1)
```

## Games ordered by average review score in the frequent dataset.

Going from highest average review to lowest.

```python
agg = frequent.groupby('gameID',as_index=False).agg(np.mean)
agg = agg.set_index('gameID').join(titles.set_index('gameID'))
agg.sort_values('rating',ascending=False)
```

## Review Distributions in Elite Dataset

The top list is the games that are most reviewed in the elite dataset. The next list is the games that have been reviewed by the smallest number of elite users. I'm a little surprised to see a Munchkin game at the bottom of the list, but since there's a million munchkin games, maybe that's to be expected. 

The graph on the left shows the distribution of how many users left a given number of reviews. The lowest number of reviews for a user in the elite list is 268, the highest is 383 out of 402 games (no one's reviewed all of them). 

Out of the 199 users in the elite dataset, the game that got reviwed the most frequently had 197 reviews. The game the least people reviewed got 42 reviews. 


```python
from IPython.display import display

def get_axis_for_histogramming():   
    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(14,7))
    ax1.set_title('how many users gave x many reviews?')
    ax1.set_xlabel('number of reviews given by a user')
    ax1.set_ylabel('number of users in bin')
    ax2.set_title('how many reviews did each game get?')
    ax2.set_xlabel('number of times a game was reviewed')
    ax2.set_ylabel('number of games in bin')
    return fig, ax1, ax2
    
fig, ax1, ax2 = get_axis_for_histogramming()
artists1 = elite.UserID.value_counts().hist(ax=ax1)
artists2 = elite.gameID.value_counts().hist(ax=ax2)
games_by_number_of_reviews = elite.gameID.value_counts().to_frame().join(titles.set_index('gameID'))
display(games_by_number_of_reviews.head(10), games_by_number_of_reviews.tail(10))
```

## Review Distributions in Frequent Dataset

The breakdown is the same as above, except from the frequent dataset. No love for Munchkin 2 or space hulk from any demographic it seems. Breakdown of review count distribution is more of a normal distribution, which is interesting since the elite users distribution had no positive tail at all. That group had a large number of games that basically everyone reviewed, while in this set, there's a much smaller number of games with near full coverage. 

```python
fig, ax1, ax2 = get_axis_for_histogramming()

frequent.UserID.value_counts().hist(ax=ax1,figure=fig)
frequent.gameID.value_counts().hist(ax=ax2,figure=fig)
games_by_number_of_reviews = frequent.gameID.value_counts().to_frame().join(titles.set_index('gameID'))
display(games_by_number_of_reviews.head(10), games_by_number_of_reviews.tail(10))
```

In the full dataset, the majory of games have few reviews, and the majority of users have left few reviews. More mainstream games are the most reviewed in this full dataset (catan, carcassone, pandemic) while in the frequent and elite most reviewed games are progressively more 'insider'.

```python
fig, ax1, ax2 = get_axis_for_histogramming()

users.UserID.value_counts().hist(ax=ax1,figure=fig)
users.gameID.value_counts().hist(ax=ax2,figure=fig)
games_by_number_of_reviews = users.gameID.value_counts().to_frame().join(titles.set_index('gameID'))
display(games_by_number_of_reviews.head(10), games_by_number_of_reviews.tail(10))
```
